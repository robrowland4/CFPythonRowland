```python
df = orders_products_merged[:1000000]
```

```python
# Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

# 03: Task

```python
#Change column to busiest days from busiest day

result = []

for value in orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 1:
    result.append("Busiest days")
  elif value == 3:
    result.append("Least busiest days")
  elif value == 4:
    result.append("Least busiest days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
orders_products_merged['busiest_days'] = result
```

```python
orders_products_merged['busiest_days'].value_counts(dropna = False)
```

"Busiest" and "least busiest" both increased, showing that the additional values were correctly encorporated.

```python
orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create new column containing labels for busiest period of day

new_result = []

for value in orders_products_merged["order_hour_of_day"]:
    if value > 7 & value <= 18 :
      new_result.append("Most_orders")
    elif value <=6:
      new_result.append("Fewest_orders")
    else:
      new_result.append("Average_orders")
```

```python
new_result
```

```python
orders_products_merged['busiest_period_of_day'] = new_result
```

```python
# Frequency for column busiest_period_of_day

orders_products_merged['busiest_period_of_day'].value_counts(dropna = False)
```
